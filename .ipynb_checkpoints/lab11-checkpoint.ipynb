## Zadanie 1. Testowanie generatorów liczb losowych

```python
import random
import numpy.random as npr
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_style('darkgrid')
rg = npr.default_rng()
```

Poniższa funkcja implementuje test szęstotliwościowy, gdzie porównujemy rozkład liczb na określonej liczbie przedziałow i wykonujemy test chi kwadrat.

Do przeprowadzania badań przyjęto poziom istotności na poziomie 1%, jest to dopuszczalne ryzyko popełnienia błędu rodzaju I, czyli uznania prawdziwej hipotezy zerowej za fałszywą. 


`pvalue` - jaka jest szansa że zobaczylibyśmy wyniki jakie zaobserwowaliśmy, gdyby liczby były na prawdę losowe.
Możemy uznać że liczby są losowe gdy pvalue < 1%



```python
from scipy.stats import chisquare
```

```python
def frequency_test(nums, bins=10):
    expected = len(nums) / bins
    freq = np.histogram(nums, bins=bins)[0]
    
    #print(np.sum(np.square(freq - expected)/expected)) - wartość statystyki
    stat, pvalue = chisquare(freq, expected)
    if pvalue < 0.01:
        print("[FAILURE] Numbers not random")
    else:
        print("[SUCCESS] Numbers random")
    print(f"pvalue = {pvalue}\n")
```

Poniższa funkcja rysuje histogram i wykonuje testy

```python
def test(nums, c='b'):
    print(f"n = {len(nums)}")
    plt.hist(nums, bins=10, color=c)
    plt.show()
    
    count = 0
    for i in range(len(nums)-1):
        if nums[i] < nums[i+1]:
            count += 1
    print(f"Count of xi < xi+1 : {count}\n") 
    
    frequency_test(nums,bins=20)
```

```python
def compare(n):
    pcg = rg.random(size=n)
    mt = [random.random() for i in range(n)]
    
    print("PCG64:")
    test(pcg)
    print()
    
    print("Mersenne Twister:")
    test(mt, c='red')
```

```python
for n in [10, 1000, 5000, 10000]:
    compare(n)
```

Jak dla każdej liczby próbek pvalue jest znacznie większe niż 1% z czego wynika że liczby możemy uznać za losowe.
Dla metody Mersenne Twister'a pvalue jest większe, ale może to być przypadek, trzeba by było wykonać więcej takich porównań.

# 2. Generowanie liczb z rozkładu normalnego

Implementacja generatora na podstawie metody Boksa-Mullera

```python
def box_muller(mu, sigma, epsilon=1e-10):
    u1 = random.random()
    u2 = random.random()
    z = np.sqrt(-2.0 * np.log(u1)) * np.cos(np.pi * 2 * u2)
    
    return z * sigma + mu
```

Rysuje histogram dla 10 przedziałów i gęstość rozkładu normalego dla wybranych parametrów mu i sigma

```python
from scipy.stats import norm
```

```python
mu = 0
sigma = 0.5

for n in [10, 1000, 5000, 10000]:
    
    nums = [box_muller(mu, sigma) for i in range(n)]
    print(f"n = {len(nums)}")

    # density normalizuje histogram żeby zmieścił się na wykresie
    # rozkład normalny
    plt.hist(nums, bins=10, density=True)
    
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, sigma)
    plt.plot(x, p)
    plt.show()
```

Jak widać liczby losowe rozkładają się zgodnie z rozkładem normalnych co widać na wykresie (zwłaszcza gdy weźmie się większą liczbę przedziałów niż 10).

# 3. Metoda Monte Carlo

```python
def sample_points(count, d=2):
    points = np.random.uniform(-1,1,(count,d))
    radii = np.sqrt(np.sum(np.square(points),axis=1)) 
    inner = points[radii < 1]
    outer = points[radii >= 1]
    return inner, outer
```

```python
def approx_pi_mc(inner, outer):
    pi_approx = 4 * len(inner)/(len(inner) + len(outer))
    return pi_approx
```

```python
# PI = 3.14159265359
inner, outer = sample_points(10**6)

pi = approx_pi_mc(inner, outer)
print(f"Approximation, PI={pi}")

plt.gca().set_aspect('equal', adjustable='box')
plt.scatter(inner[:,0],inner[:,1])
plt.scatter(outer[:,0],outer[:,1])
plt.show()
```

```python
inner, outer = sample_points(10**7)

pi = approx_pi_mc(inner, outer)
print(f"Approximation, PI={pi}")
```

Jak widać korzystając z metody Monte Carlo można uzyskać przybliżenie liczby PI, chociaż ciężko jest uzyskać dokładność do 3 miejsc po przecinku, gdyż wymaga to losowania dużej ilości punktów ~10^7

Przygotowanie danych pod wykres błędy przybliżenia od liczby punktów.

```python
ns = np.linspace(5000, 10**6, num=200, dtype=np.int32)
pies = np.empty(ns.shape[0])
for i, n in enumerate(ns):
    inner, outer = sample_points(n)
    pies[i] = approx_pi_mc(inner, outer)
```

```python
plt.scatter(ns, np.abs(pies - np.pi))
plt.ylim(bottom=-1e-4)
plt.show()
```

Jak widać wraz ze wzrostem liczby losowanych punktów maleje błąd przybliżenia PI. Metoda Monte Carlo jest pożyteczna ale jest mało efektywana. Można ją wykorzystać np do obliczania całek.
