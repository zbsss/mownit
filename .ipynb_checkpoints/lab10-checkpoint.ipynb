## Zadanie 1. FFT
### 1. 
Zaimplementuj funkcję realizującą DFT jako iloczyn macierzy

```python
def DFT_matrix(x):
    N = x.shape[0]
    M = np.array([[-2j * np.pi * k * n / N for n in range(N)] for k in range(N)])
    M = np.exp(M)
    return M
```

```python
def DFT_simple(x):
    M = DFT_matrix(x)
    return M @ x
```

### 2. 
Sprawdź poprawność działania funkcji realizującej DFT stosując transformację odwrotną oraz porównując uzyskane wyniki z wyjściem funkcji bibliotecznej.


Porównanie wyników mojej implementacji z wynikami funkcji bibliotecznej

```python
x = np.random.random(1024)
np.allclose(DFT_simple(x), np.fft.fft(x))
```

```python
x = np.random.random(2048)
np.allclose(DFT_simple(x), np.fft.fft(x))
```

Jak widać wyniki są zgodne

Następnie przygotowałem funkcje do obliczania transformacji odwrotnej

```python
def IDFT_simple(x):
    M = DFT_matrix(x)
    y = DFT_simple(x)
    return np.linalg.inv(M) @ y
```

```python
x = np.random.random(1024)
np.allclose(x, IDFT_simple(x))
```

Jak widać wyniki są zgodne, program działa zgodnie z wzorami

## 3.
Zaimplementuj rekurencyjny algorytm Cooleya-Turkeya realizujący szybką transformację Fouriera (FFT). Porównaj szybkość jego działania z implementacją biblioteczną oraz implementacją z mnożeniem wektora przez macierz Fn dla danych
o różnym rozmiarze.


```python
def FFT(x):
    N = len(x)
    
    if N <= 1: 
        return x
    
    y_even = FFT(x[0::2])
    y_odd =  FFT(x[1::2])
    
    T = [np.exp(-2j * np.pi * k / N) * y_odd[k] for k in range(N//2)]
    return [y_even[k] + T[k] for k in range(N//2)] +[y_even[k] - T[k] for k in range(N//2)]
```

```python
x = np.random.random(1024)
np.allclose(FFT(x), np.fft.fft(x))
```

Jak widać wyniki mojej implementacji są poprawne

```python
from time import time

def timer(func, *args):
    start = time()
    func(*args)
    end = time()
    return end - start
```

Następnie zmierzyłem czasy dla wszystkich implementacji

```python
for i in range(13):
    x = np.random.random(2**i)
    print("-----------------------------------")
    print(f"N = {2**i}")
    print(f"Simple DFT: {timer(DFT_simple, x)}")
    print(f"FFT my implementation: {timer(FFT, x)}")
    print(f"FFT from numpy: {timer(np.fft.fft, x)}")    
```

Jak widać prosta implementacja macierzowa jest znacząco wolniejsza od szybkich algorytmów FFT. Implementacja z biblioteki numpy niewzruszona.

# Zadanie 2. DFT w 1D
Wygeneruj dwa sygnały czasowo-amplitudowe:


```python
f_s = 100 # liczba pomiarów na sekunde
t = np.linspace(0, 2, 2 * f_s) # czas
signal = lambda t, f : np.sin(2 * np.pi * t * f)
```


#### a) Sygnał będący sumą pięciu sygnałów sinusoidalnych o różnych częstotliwościach

Generuje sygnały sinusoidalne o różnych częstotliwościach, poniżej wykres każdego z nich, a następnie sumuje do jednego sygnału.

```python
sum_signal = 0
for f in [200, 500, 1000, 3000, 5000]:
    s = signal(t, f)
    sum_signal += s
    plt.plot(t, s)
    plt.show()
```

Wykres sygnału będącego sumą poprzenich pięciu

```python
plt.plot(t, sum_signal)
plt.show()
```

Następnie dokonuje transformacji sygnału do domeny częstotliwości

```python
X = np.fft.rfft(sum_signal)
f = np.fft.rfftfreq(len(sum_signal)) * f_s
plt.plot(f,np.abs(X))
plt.show()
```

#### b) Sygnał złożony z pięciu sygnałów o tych samych częstotliwościach co w punkcie a), ale ułożonych przedziałami, tzn. w każdym z pięciu przedziałów o tej samej szerokości występuje sygnał o jednej częstotliwości

Tworzę syngał posklejany z pięciu sygnałów z poprzedniego podpunktu, poniżej jego rysunek

```python
F = [200, 500, 1000, 3000, 5000]
stitched_signal = np.array([signal(t[i:i+len(t)//5], F[i]) for i in range(5)]).flatten()
plt.plot(t, stitched_signal)
plt.show()
```

Jak widać sygnał wydaje się być zgodny z wykresami z poprzeniego podpunktu

Dokonuje transformacji sygnału do domeny częstotliwości

```python
X = np.fft.rfft(stitched_signal)
f = np.fft.rfftfreq(len(stitched_signal)) * f_s
plt.plot(f,np.abs(X))
plt.show()
```

Wykresy dla podpunktu a i b są znacząco różne, chociaż widać "piki" w tych samych miejsach
