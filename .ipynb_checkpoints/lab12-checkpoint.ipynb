### 1. Opisz zalety metody Rungego-Kutty w porównaniu do metody z szeregami Taylora
- **Metoda Eulera** jest mało precyzyjna, błąd rzędu O(h^2) dla dwóch wyrazów szeregu Taylora. Trzeba znać pochodne funkcji żeby liczbyć kolejne wyrazy szeregu Taylora co nie zawsze jest łatwe oraz wymaga dodatkowej pracy przed uruchomieniem programu.

- **Metoda Rungego-Kutty** nie wymaga liczenia pochodnych. Przez co jej używanie jest znacząco prostsze, wystarczy uruchomić algorytm dla danej funkcji f(t,x).

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 2.

Moja impementacja motody Rungego-Kutty:

```python
def runge_kutta(f, a, b, x, h):
        t = a
        X = []
        T = []
        
        n = int((b - a)/h)
        
        for j in range(1, n+1):
            K1 = h * f(t, x)
            K2 = h * f(t + 1/2 * h, x + 1/2 * K1)
            K3 = h * f(t + 1/2 * h, x + 1/2 * K2)
            K4 = h * f(t + h, x + K3)

            x = x + 1/6 * (K1 + 2*K2 + 2*K3 + K4)
            t = a + j * h
            j += 1
            
            T.append(t)
            X.append(x)
        return T, X
```

```python
f = lambda t, x: x/t + t * 1/np.cos(x/t)
```

```python
ts, xs = runge_kutta(f, x=0, a=1e-12, b=1, h=2**-7)
```

Metoda Rungego-Kutty:

```python
plt.plot(ts, xs)
plt.show()
```

Dokładne rozwiązanie:

```python
plt.plot(ts, ts * np.arcsin(ts))
plt.show()
```

Oba na tym samym wykresie:

```python
plt.plot(ts, xs)
plt.plot(ts, ts * np.arcsin(ts))
plt.show()
```

Jak widać na wykresie rozwiązanie z metody Rungego-Kutty pokrywa się z rozwiązaniem dokładnym.

# 2. Adaptacyjna metoda Rungego-Kutty-Fehlberga

```python
def runge_kutta_fehlberg(f, a, b, x, h, tolerance, max_inter=10000):
    c20 = c21 = 0.25
    c30 = 0.375
    c31 = 0.09375
    c32 = 0.28125
    c40 = 12/13
    c41 = 1932/2197
    c42 = -7200/2197
    c43 = 7296/2197
    c51 = 439/216
    c52 = -8
    c53 = 3680/513
    c54 = -845/4104
    c60 = 0.5
    c61 = -8/27
    c62 = 2
    c63 = -3544/2565
    c64 = 1859/4104
    c65 = -0.275
    a1 = 25/216
    a2 = 0
    a3 = 1408/2565
    a4 = 2197/4104
    a5 = -0.2
    b1 = 16/135
    b2 = 0
    b3 = 6656/12825
    b4 = 28561/56430
    b5 = -0.18
    b6 = 2/55
    t = a
    
    X = []
    T = []
    stop = False
    
    for i in range(max_inter):
        K1 = h * f(t, x)
        K2 = h * f(t + c20 * h, x + c21 * K1)
        K3 = h * f(t + c30 * h, x + c31 * K1 + c32 * K2)
        K4 = h * f(t + c40 * h, x + c41 * K1 + c42 * K2 + c43 * K3)
        K5 = h * f(t + h, x + c51 * K1 + c52 * K2 + c53 * K3 + c54 * K4)
        K6 = h * f(t + c60 * h, x + c61 * K1 + c62 * K2 + c63 * K3 + c64 * K4 + c65 * K5)

        x4 = x + a1 * K1 + a3 * K3 + a4 * K4 + a5 * K5
        x5 = x + b1 * K1 + b3 * K3 + b4 * K4 + b5 * K5 + b6 * K6

        eps = abs(x5 - x4)
    
        # jeśli błąd jest niższy niż tolerancja akceptujemy wynik
        if eps <= tolerance:
            t = t + h
            x = x4
            
            T.append(t)
            X.append(x)
            if stop:
                break
        
        # ustawienie następnego kroku
        if eps != 0:
            # z książki "Numerical Methods in Engineering with Python"
            h_next = 0.9 * h * (tolerance/eps) ** 0.2
        else:
            h_next = h
            
        # jeśli następny krok jest ostatnim
        if (h > 0.0 and t + h_next >= b) or (h < 0.0 and t + h_next <= b):
            h_next = b - t
            stop = True
            
        h = h_next
    return T, X
```

```python
f = lambda t, x: 3*x/t + 9/2 * t - 13
t, x = runge_kutta_fehlberg(f, a=3, b=1/2, x=6, h=-0.1, tolerance=1e-10)
```

Wykres otrzymany na podstawie metody Adaptacyjnej Rungego-Kutty-Fehlberga:

```python
plt.plot(t,x)
plt.show()
```

Wykres rozwiązania analitycznego:

```python
ts = np.linspace(1/2, 3, 1000)
sol = lambda t: 1/2 * t * (2 * t**2 - 9 * t + 13)
plt.plot(ts, list(map(sol, ts)))
plt.show()
```

Końcowy wynik otrzymany na podstawie metody Adaptacyjnej Rungego-Kutty-Fehlberga:

```python
print(x[-1])
```

Końcowy wynik rozwiązania analitycznego:

```python
print(sol(1/2))
```

Błąd bezwzględny:

```python
print(abs(x[-1] - sol(1/2)))
```

Jak wiać wynik otrzymany na podstawie metody adaptacyjnej Rungego-Kutty-Fehlberga jest zgodny z wynikiem rozwiązania analitycznego z dokładnością do 9 miejsc po przecinku.

Metody Adaptacyjna umożliwia zmienianie kroku h zależnie od tego czy błąd obliczeń jest większy czy mniejszy od tego jaki chcieliśmy uzyskać. Kiedy błąd jest zbyt duży nie zapisujemy wyniku tylko wracamy do poprzedniego i zmiejszamy długość kroku. Dzięki temu przejście z x(t) do x(t+h) możemy rozbić na x(t) -> x(t+h/2) -> x(t+h) co wpływa na zmiejszenie błedu.
Kiedy błąd jest mniejszy niż oczekiwaliśmy możemy zwiększyć długośc kroku z przez to zmiejszyć liczbę iteracji potrzebną do obliczenia wyniku

Wadą może być to że procedura adaptująca moze się zapętlić próbujac na zmianę skrócić i przedłużyć przedział o 2. Można temu zapobiec ustawiając maksymalne i minimalne limity na długość kroku i wielkość błędu.
