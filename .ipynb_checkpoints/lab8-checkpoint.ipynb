W ramach testu tworze losowy graf silnie spójny:

Obliczam Page Rank za pomocom funkcji bibliotecznej do późniejszego porównania z moimi wynikami:

```python
PRankH = snap.TIntFltH()
snap.GetPageRank(G, PRankH)
```

```python
for item in PRankH:
    print(item, PRankH[item])
```

Przygotowuje przeskalowaną macierz adiacencji A:

```python
def prep_A(G):
    n = G.GetNodes()

    N = np.ones((n))
    for node in G.Nodes():
        N[node.GetId()] = node.GetOutDeg()


    A = np.zeros((n, n))
    for edge in G.Edges():
        u, v = edge.GetId()
        A[u][v] = 1/N[u]

    A = A.T
    return A
```

```python
A = prep_A(G)
print(A)
```

Do obliczania dominującego wektora własnego wykorzystuje swoją implementacje metody potęgowej z poprzedniego laboratorium.

```python
def power_method(A, max_iterations=10000, epsilon=1e-8):
    n = A.shape[0]
    x = np.random.uniform(size=(n, 1))
    
    i = 0
    while i < max_iterations:
        x_i = A @ x
        x_i /= np.linalg.norm(x_i)
        
        if np.linalg.norm(x_i - x) < epsilon:
            break
        
        x = x_i
        i += 1
        
    return x.T @ A @ x_i, x / np.linalg.norm(x), i
```

Obliczam page rank korzystając z dominującego wektora własnego macierzy A:

```python
def page_rank(A):
    _, r, _ = power_method(A)

    c  = 1 / np.sum(r)
    r =  c * r
    
    return r
```

Funkcja do porównywania wyników mojej implementacji page ranka i wyniku funkcji bibliotecznej:

```python
def compare_results(r, epsilon=1e-4):
    for item in PRankH:
        print(r[item], PRankH[item], end=" ")
        if abs(PRankH[item] - r[item]) < epsilon:
            print("SUCCESS")
        else:
            print("FAIL")
```

```python
r = page_rank(A)
compare_results(r, epsilon=1e-3)
```

```python
compare_results(r, epsilon=1e-4)
```

Jak widać dla epsilon = 1e-3 wszystkie wyniki są zgodne, dla 1e-4 zaczynają pojawiać się błędy

## Zadanie 2 Page Rank
Teraz do prostej implementacji rankingu wierzchołków, dodaje możliwość skoku do dowolnego wierzchołka w grafie. Za ten losowy skon będzie odpowiadał wektor e prawdopodobieństw skoku do danego wierzchołka.

Implementacja Page Rank ze skokami:

```python
def page_rank_with_jumps(A, e, d=0.9):
    B = d * A + (1-d)*e
    _, r, _ = power_method(B)

    c  = 1 / np.sum(r)
    r =  c * r

    return r
```

```python
e = 1/n * np.ones(n)
R = page_rank_with_jumps(A, e)
compare_results(R, epsilon=1e-4)
```

Jak widać wszystkie wyniki są zgodne dla epsilon=1e-4, przy takim błędzie prosta implementacja generowała już sporo błędów. Implementacja Page Rank z losowymi skokami jest bliższa implementacji bibliotecznej.

Wyniki dla różnych parametrów d:

```python
ds = [0.9, 0.85, 0.75, 0.6, 0.5]
for d in ds:
    e = 1/n * np.ones(n)
    
    print("\nd :", d)
    print("e :\n",e)
    
    R = page_rank_with_jumps(A, e, d=d)
    compare_results(R, epsilon=1e-4)
```

Jak widać dla małych wartości parametru d zaczyna pojawiać się coraz więcej błedów.

Wyniki dla losowo wybranych wektorów e:

```python
for i in range(10):
    e = np.random.rand(n)
    e *= 1/np.sum(e)
    
    print("e :\n",e)
    
    R = page_rank_with_jumps(A, e)
    compare_results(R, epsilon=1e-4)
```

Jak wiać wektor e nie ma zbyd dużego wpływu na poprawność wyniku, większe znaczenia ma wartość parametru d.
