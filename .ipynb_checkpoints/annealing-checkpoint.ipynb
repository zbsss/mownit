```python
import numpy as np
import matplotlib.pyplot as plt
from random import random
```

Za rozwiązanie przyjąłem ciąg kolejnych punktów w 2D, więc zmieniając ich kolejność w liście zmieniam rozwiązanie.

```python
def anneal(sol):
    old_cost = cost(sol)
    T = 1.0
    T_min = 0.0000001
    alpha = 0.9
    while T > T_min:
        i = 1
        while i <= 100:
            new_sol = neighbor(sol)
            new_cost = cost(new_sol)
            ap = acceptance_probability(old_cost, new_cost, T)
            if ap > random():
                sol = new_sol
                old_cost = new_cost
            i += 1
        T = T*alpha
    return sol, new_cost
```

```python
def cost(solution):
    cost = 0
    n = len(solution)
    for i in range(n):
        cost += np.linalg.norm(solution[i] - solution[(i+1) % n])
    return cost
```

```python
def neighbor(sol):
    solution = np.copy(sol)
    x, y = np.random.randint(low=0, high=len(solution), size=2)
    solution[[x, y]] = solution[[y, x]]
    return solution
```

$a=e^{\frac{c_{n e w}}{T}-c_{o l d}}$

```python
def acceptance_probability(cost, new_cost, temperature):
    if new_cost < cost:
        # print("    - Acceptance probabilty = 1 as new_cost = {} < cost = {}...".format(new_cost, cost))
        return 1
    else:
        p = np.exp(- (new_cost - cost) / temperature)
        # print("    - Acceptance probabilty = {:.3g}...".format(p))
        return p
```

```python
n = 20
points = np.random.random(size=(n, 2))
plt.scatter(points[:, 0], points[:, 1], c='red')
plt.plot(points[:, 0], points[:, 1])
plt.show()
```

```python
sol, co = anneal(points)
```

```python
# dodaje kopie pierwszego punktu na koniec żeby narysować cykl a nie ścieżkę
sol = np.concatenate((sol, sol[0,None]), axis=0)

plt.scatter(sol[:, 0], sol[:, 1], c='red')
plt.plot(sol[:, 0], sol[:, 1])
plt.show()
```
