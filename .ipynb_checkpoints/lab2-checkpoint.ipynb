```python
import numpy as np
import matplotlib.pyplot as plt
```

### Zad 1
#### a)

```python
def f(x):
    return 1/(1 + x**2)

def p(x, a, n):
    return np.dot(np.vander(x, N=n+1, increasing=True), a)

def interpolate(n):
    knots = np.linspace(-5, 5, n + 1)
    vander = np.vander(knots, increasing=True)
    y = np.array([f(i) for i in knots])
    
    return np.linalg.solve(vander,y)
    
def interpolation_plot(n):
    x = np.linspace(-5, 5, 1000)

    a = interpolate(n)

    plt.plot(x, p(x, a, n))
    plt.plot(x, f(x))
    plt.show()
```

```python
interpolation_plot(5)
```

```python
interpolation_plot(10)
```

```python
interpolation_plot(15)
```

#### b)

```python
def interpolation_remainder_plot(n):
    x = np.linspace(-5, 5, 1000)

    a = interpolate(n)

    remainder = [f(x[i]) - p(x, a, n)[i] for i in range(len(x))]
    
    plt.plot(x, p(x, a, n))
    plt.plot(x, f(x))
    plt.plot(x, remainder)
    plt.show()
```

```python
interpolation_remainder_plot(15)
```
