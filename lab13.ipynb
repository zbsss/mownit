```python
import time

def timeit(func):
    def timed(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        
        print(f"time: {end - start}")
        
        return result
    return timed
```

```python
def cost(solution):
    cost = 0
    n = len(solution)
    for i in range(n):
        cost += np.linalg.norm(solution[i] - solution[(i+1) % n])
    return cost
```

Zamieniam dwa losowo wybrane elementy w ścieżce:

```python
def arbitrary_swap(sol):
    solution = np.copy(sol)
    
    x = y = 0
    while x == y:
        x, y = np.random.randint(low=0, high=len(solution), size=2)

    # zamiana dwóch elementów w ścieżce
    solution[[x, y]] = solution[[y, x]]
    return solution
```

Zamieniam dwa miejscami dwa kolejne elementy w ścieżce:

```python
def consecutive_swap(sol):
    solution = np.copy(sol)
    
    x = np.random.randint(low=0, high=len(solution)-1)
    y = x + 1

    # zamiana dwóch elementów w ścieżce
    solution[[x, y]] = solution[[y, x]]
    return solution
```

Funkcja generująca prawdopodobieństwo przyjęcia zamiany punktów na ścieżce :
$a=e^{\frac{c_{n e w}}{T}-c_{o l d}}$

```python
def acceptance_probability(cost, new_cost, temperature):
    return 1 if new_cost < cost else np.exp(- (new_cost - cost) / temperature)
```

Implementacja algorytmu symulowanego wyżarzania do przybliżonego rozwiązania problemu komiwojażera:

```python
@timeit
def anneal(sol, T = 1.0, T_min = 0.0000001, alpha = 0.9, neighbor=arbitrary_swap):
    old_cost = cost(sol)
    
    while T > T_min:
        i = 1
        while i <= 100:
            new_sol = neighbor(sol)
            new_cost = cost(new_sol)
            ap = acceptance_probability(old_cost, new_cost, T)
            if ap > random():
                sol = new_sol
                old_cost = new_cost
            i += 1
        T = T * alpha
    return sol, new_cost
```

Generuje losowe punkty i rysuje pierwszą losową ścieżkę:

```python
def anneal_save(sol, T = 1.0, T_min = 0.0000001, alpha = 0.9, neighbor=arbitrary_swap):
    old_cost = cost(sol)
    
    while T > T_min:
        i = 1
        while i <= 100:
            new_sol = neighbor(sol)
            new_cost = cost(new_sol)
            ap = acceptance_probability(old_cost, new_cost, T)
            if ap > random():
                sol = new_sol
                old_cost = new_cost
                history.append(sol)
            i += 1
        T = T * alpha
    return sol, new_cost
```

```python
history = []

points = np.random.random(size=(40, 2))
sol, co = anneal_save(np.copy(points))
print(f"Cost: {co}")

sol = np.concatenate((sol, sol[0,None]), axis=0)
print(len(history))

plt.scatter(sol[:, 0], sol[:, 1], c='red')
plt.plot(sol[:, 0], sol[:, 1])
plt.show()
```

```python
import matplotlib.animation as animation
from IPython.display import HTML

fig = plt.figure()
ax = plt.axes()
line, = ax.plot([], [], lw=2) 

def init(): 
    line.set_data([], []) 
    return line, 

def animate(i):
    xs = history[i][:, 0]
    ys = history[i][:, 1]
    
    ax.clear()
    
    plt.scatter(xs, ys, c='red')
    plt.plot(xs, ys)
    
    line.set_data(xs, ys)
    return line,

anim = animation.FuncAnimation(fig, animate, interval=100, save_count=len(history)) 

HTML(anim.to_html5_video())
```
